# Forward Propagation and Backward Propagation

1. Forward Propagation

👉 Definition:
Forward propagation is the process where input data passes through the neural network to produce an output (prediction).

🧠 Flow:
Input → Hidden Layers → Output

3. Backward Propagation

👉 Definition:
Backward propagation is the process of sending the error (loss) backward through the network to update weights.

🧠 Flow:
Output → Hidden Layers → Input (reverse)

What happens:

Compute gradients (how much error depends on each weight)
Update weights
Reduce loss
