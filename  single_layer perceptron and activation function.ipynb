```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\Data_set\logistic_regression_student_placement.csv")
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test)*100,pr.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
```
