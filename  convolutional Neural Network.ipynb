# convolutional Neural Network

```python
import tensorflow
```

```python
from keras.layers import Dense , Conv2D , MaxPooling2D ,Flatten
from keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
cnn = Sequential()
```

```python
cnn.add(Conv2D(32,(3,3),input_shape = (64,64,3),activation="relu")) # filter first=32,filter size=3,3,input_size=64*64*3
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(16,(3,3),activation="relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Flatten())
```

```python
cnn.add(Dense(64,activation="relu"))
cnn.add(Dense(32,activation="relu"))
cnn.add(Dense(16,activation="relu"))
cnn.add(Dense(8,activation="relu"))
cnn.add(Dense(4,activation="relu"))
cnn.add(Dense(2,activation="relu"))
cnn.add(Dense(1,activation="sigmoid"))
```

```python
cnn.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\Data_set\dogs_vs_cats\train",
        target_size=(64,64),
        batch_size=32,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        r"C:\Users\imrat\OneDrive\Desktop\data_science_all_uses_items\Data_set\dogs_vs_cats\test",
        target_size=(64,64),
        batch_size=32,
        class_mode='binary')

cnn.fit(train_generator,steps_per_epoch=20,epochs=5,validation_data=test_generator)
```
